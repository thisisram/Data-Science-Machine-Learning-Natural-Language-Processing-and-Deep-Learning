```python
digits = pd.read_csv('../Data/digit_recognizer_train.csv')
```

```python
digits.columns
```

```python
X = digits.iloc[:, 1:]
y = digits.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train = X_train.astype(np.float64)
X_test = X_test.astype(np.float64)
```

```python
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import BatchNormalization

from tensorflow.keras import Model
from tensorflow.keras import Input
from tensorflow.keras.layers import Dense
```

```python
training=True

x_in = Input(shape=(784,))

layer1 = Dense(128)(x_in)
norm_layer1 = BatchNormalization()(layer1, training)
relu_layer1 = ReLU()(norm_layer1)

layer2 = Dense(128)(relu_layer1)
norm_layer2 = BatchNormalization()(layer2, training)
relu_layer2 = ReLU()(norm_layer2)

#layer3 = Dense(128)(relu_layer2)
#norm_layer3 = BatchNormalization()(layer3, training)
#relu_layer3 = ReLU()(norm_layer3)

#layer4 = Dense(128)(relu_layer3)
#norm_layer4 = BatchNormalization()(layer4, training)
#relu_layer4 = ReLU()(norm_layer4)

out_layer =  Dense(10)(relu_layer2)
x_out = BatchNormalization()(out_layer, training)

model = Model(inputs=x_in, outputs=x_out)
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
sgd = SGD(learning_rate=0.001, decay=0.0001)
model.compile(optimizer=sgd, loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train.values, y_train.values,  batch_size=512, epochs = 300, verbose=1)
```

```python
model.summary()
```

```python
model.evaluate(X_test.values, y_test.values)   #Output:[Loss, accuracy_score]
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_test)
```
